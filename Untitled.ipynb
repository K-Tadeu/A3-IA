```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

# Carregar dados e verificar a distribuição original

```python
data = pd.read_csv("mushroom_cleaned.csv")
```

# Verificar Balanceamento de Classes

```python
print(data['class'].value_counts())

sns.countplot(x='class', data=data)
plt.title('Distribuição da Classe (0 = Comestível , 1 = Venenoso)')
plt.show()

value_counts = data['class'].value_counts()

max_class = max(value_counts[1], value_counts[0])
min_class = min(value_counts[1], value_counts[0])

desbalanced_percentage = (max_class - min_class) / (max_class+min_class) * 100
print("Desbalanceamento de ", round(desbalanced_percentage,2), '%', sep='')
```

# Preparar variáveis independentes e dependentes

```python
X = data.drop('class', axis=1)
y = data['class']
```

# Aplicar SMOTE para balancear as classes

```python
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
data_resampled = pd.DataFrame(X_res, columns=X.columns)
data_resampled['class'] = y_res


sns.countplot(x='class', data=data_resampled)
plt.title('Distribuição da Classe (0 = Comestível , 1 = Venenoso)')
plt.show()

correlation = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, cmap="coolwarm", annot=False, fmt='.2f')
plt.title("Mapa de Correlação das Variáveis do Dataset")
plt.show()




```

# Função para calcular e exibir métricas de avaliação

```python
# Função para calcular e exibir métricas de avaliação
def evaluate_model(X, y, model):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred

    # Matriz de Confusão
    cm = confusion_matrix(y_test, y_pred)
    print(f"\nMétricas para {type(model).__name__}")
    print("Matriz de Confusão:")
    print(cm)

    # Relatório de Classificação
    report = classification_report(y_test, y_pred, output_dict=True)
    print(classification_report(y_test, y_pred))

    # AUC-ROC
    roc_auc = roc_auc_score(y_test, y_proba)

    # F1-score para a classe minoritária
    f1_minor_class = report["1"]["f1-score"]

    return roc_auc, f1_minor_class, model, X_test, y_test
```

# Modelos a serem usados no comitê

```python
models = [RandomForestClassifier(), KNeighborsClassifier(n_neighbors=3), DecisionTreeClassifier()]
results = {}
```

# Função Comitê de IAs

```python
def comite(X, y, models, datasetName):
    # Avaliar cada modelo e armazenar os resultados
    for model in models:
        roc_auc, f1_score_minor, fitted_model, X_test, y_test = evaluate_model(X, y, model)
        results[type(model).__name__] = {
            "AUC-ROC": roc_auc,
            "F1-score Minor Class": f1_score_minor,
            "model": fitted_model
        }

    # Calcular pesos com base no desempenho dos modelos
    total_auc = sum(result["AUC-ROC"] for result in results.values())
    for model_name in results:
        results[model_name]["weight"] = results[model_name]["AUC-ROC"] / total_auc

    # Predições ponderadas
    final_proba = np.zeros(len(y_test))
    for model_name, result in results.items():
        weight = result["weight"]
        model = result["model"]
        proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else model.predict(X_test)
        final_proba += weight * proba
    print(final_proba)

    # Converter probabilidades finais em predições finais
    final_pred = [1 if p > 0.5 else 0 for p in final_proba]

    # Matriz de Confusão e Relatório de Classificação do Comitê
    cm = confusion_matrix(y_test, final_pred)

    # Gráfico da Matriz de Confusão do Comitê
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Legítima", "Fraude"], yticklabels=["Legítima", "Fraude"])
    plt.xlabel("Predição")
    plt.ylabel("Valor Real")
    plt.title(datasetName + " - Matriz de Confusão - Comitê de Classificação")
    plt.show()

    # Relatório de Classificação do Comitê
    print(datasetName + " - Relatório de Classificação do Comitê:")
    print(classification_report(y_test, final_pred))

    # Curva ROC para o Comitê de Classificação
    fpr, tpr, thresholds = roc_curve(y_test, final_proba)
    roc_auc = roc_auc_score(y_test, final_proba)
    plt.figure()
    plt.plot(fpr, tpr, label="AUC-ROC Curve (AUC = {:.2f})".format(roc_auc))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("Taxa de Falsos Positivos")
    plt.ylabel("Taxa de Verdadeiros Positivos")
    plt.title(datasetName + " - Curva ROC - Comitê de Classificação")
    plt.legend(loc="lower right")
    plt.show()

    # Criar DataFrame com as classificações
    classification_results = pd.DataFrame({
    "Índice": X_test.index,
    "Classe Esperada": y_test.values,
    "Classe Predita": final_pred
    })

    # Exibir as primeiras linhas dos resultados de classificação
    print(classification_results.head(100))
```

# Desempenho do Dataset Não-balanceado

```python
comite(X,y,models, "Dataset Não-balanceado")
```

# Desempenho do Dataset Balanceado

```python
comite(X_res, y_res, models, "Dataset Balanceado")
```
